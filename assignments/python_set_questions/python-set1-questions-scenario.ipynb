```python
def check_pass_or_fail(marks):
    return sum(1 for mark in marks if mark < 40), sum(1 for mark in marks if mark >= 40)
    
marks_list = [10, 60, 70, 35]
failed, pass_count = check_pass_or_fail(marks_list)
print(f"Number of subjects passed: {pass_count}, Number of subjects failed: {failed}.")
```

```python
def calculate_discount_total(purchased_amt):
    if purchased_amt > 100:
        percent = (( purchased_amt * 20)/100)
        total_amt = purchased_amt - percent
        return total_amt
    else:
        return total_amt

purchase_price = 200
total_cost_to_pay = calculate_discount_total(purchase_price)
print(f"Purchased price: {purchase_price}. After discount of 20%, total cost to pay: {total_cost_to_pay}" if purchase_price > 100 else f"Total cost to pay: {purchase_price}.")
```

```python
available_balance = 1000

def withdrawal(debit_amt, available_balance):
    if debit_amt > available_balance:
        return f"Available balance is {available_balance}. Cannot debit the requested amount {debit_amt}. Please try to debit less than or equal to available balance."
    elif debit_amt < available_balance:
        available_balance -= debit_amt
        return f"Requested amount {debit_amt} withdrawal successfully. Current balance {available_balance}."

debit = 999
print(withdrawal(debit, available_balance))
```

```python
def find_leap_year(year):
    if (( year % 4) == 0 and (year % 100) != 0) or (( year % 400) == 0):
        return "Leap year"
    else:
        return "Not leap year"

input_year = int(input("Enter the year to validate the leap year or not: "))
print(find_leap_year(input_year))
```

```python
def fetch_even_nums(input_list):
    even_nums = [ enums for enums in input_list if enums % 2 == 0 ]
    return even_nums

user_list = input("Enter the list of numbers to fetch only even numbers (comma separated as input): ")
user_list_new = [ int(num.strip()) for num in user_list.split(',')]
fetch_even_nums(user_list_new)
```
