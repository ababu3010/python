```python
def factorial(n, cache={}):
    if n < 0:
        raise ValueError("Negative value")
    if n in cache:
        return cache[n]
    if n == 0 or n == 1:
        cache[n] = 1
    else:
        cache[n] = n * factorial(n-1)
    return cache[n]

factorial(5)
```

```python
def check_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if (n % i) == 0:
            return False
    return True

n = 7
result = check_prime(n)
if result:
    print(f"{n} is a prime number")
else:
    print(f"{n} is not a prime number")
```
