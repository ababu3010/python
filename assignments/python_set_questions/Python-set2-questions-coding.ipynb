```python
input_integer = int(input("Enter the input integer: "))
def fibonacci(n):
    a, b = 0, 1
    for i in range(0,n):
        print(a,end=" ")
        a, b = b, a+b
        
fibonacci(input_integer)
```

```python
string = input("Enter the word: ")
reverse_string = string[::-1]
if string == reverse_string:
    print(f"{string} is Palindrom")
else:
    print(f"{string} is not palindrom")
```

```python
string_input = input("Enter the word to count vowels letter: ")
vowels = ['a','e','i','o','u']
count = 0
for letter in string_input:
    if letter in vowels:
        count += 1
print(f"Number of vowels: ", count)
```
